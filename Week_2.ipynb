```python
from sklearn.datasets import load_breast_cancer

# Load the breast cancer dataset
breast_cancer = load_breast_cancer()

# Get the feature matrix
X = breast_cancer.data

# Print the shape of the feature matrix
print(X.shape)
```

```python
import pandas as pd

df = pd.DataFrame(X)
```

```python
df
```

```python
y = breast_cancer.target

# Count the number of benign (B) tumor cases
malignant_cases = (y == 0).sum()

# Print the number of benign tumor cases
print(f"Number of malignant (B) tumor cases: {malignant_cases}")
```

```python
y = breast_cancer.target

# Count the number of benign (B) tumor cases
bengin_cases = (y == 1).sum()

# Print the number of benign tumor cases
print(f"Number of benign (B) tumor cases: {bengin_cases}")
```
